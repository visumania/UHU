# Ejercicio 

Diseñar un programa en *Python* que extraiga, a través de la API *Pixabay*, todas las imágenes que cumplan varios requisitos de búsqueda y las guarde en una carpeta como ficheros independientes. 
- Los parámetros se pasarán en la línea de comandos al hacer la llamada al programa. 
- Por ejemplo, podéis crear una función que admita un número variable de parámetros de tipo diccionario: `q='edificio', colors='blue', ...}` y que devuelve el conjunto de parámetros que se le pasan al método `requests.get()` junto con el *endpoint*.

```python
import requests
import json
```

```python
api_key="20748390-db8e97c4f753ba64d200feef2" 
url_endpoint = "https://pixabay.com/api/"
```

```python
# Recogida de información para realizar la petición a la API
per_page = 200

print("Introduzca query (escape si no se quiere introducir ninguna):")
query = input()

print("Introduzca color/es (escape si no se desea ningún color en especial):")
colores = input()

print("Introduzca orientación de la imagen: [all, horizontal, vertical]")
orientacion = input()

print("¿Deseas que las imágenes devueltas sean seguras (s/n)?")
safesearch = input().upper
seguras = False
if(safesearch=='S'):
    seguras=True

params = {'key':api_key, 'q':query, 'colors':colores, 'orientation':orientacion, 'safesearch':safesearch, 'per_page':per_page}
```

```python
respuesta = requests.get(url=url_endpoint, params=params)
datos = respuesta.json()
print(json.dumps(datos, indent=2))

url_links = []
for imagen in datos['hits']:
    url_links.append([imagen["webformatURL"], imagen["previewURL"]])
```

```python
indice = 0
for imagen in url_links:
    indice += 1
    r = requests.get(imagen[0], stream=True)

    nombre_fichero = imagen[1].split("/")[-1]

    open("./fotosPixabay/" + nombre_fichero, 'wb').write(r.content)
    
```
